```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv(r"C:\Users\Nithya\OneDrive\Documents\companies_data.csv")
data
```

```python
data.info()
```

```python
data=data.drop(["Unnamed: 0"],axis=1)
data
```

```python
data["RatedFor"]=data["RatedFor"].fillna("Not Available")
```

```python
lst1=[]
lst2=[]
for i in range(0,500):
    if data["RatedFor"][i]=='Not Available':
        lst1.append("Not Available")
        lst2.append("Not Available")
    elif "Critically Rated For" not in data["RatedFor"][i]:
        lst1.append(data["RatedFor"][i].split("For")[1].strip())
        lst2.append("Not Available")
    elif "Highly Rated For" not in data["RatedFor"][i]:
        lst1.append("Not Available")
        lst2.append(data["RatedFor"][i].split("For")[1].strip())
    else:
        lst1.append(data["RatedFor"][i].split("Critically")[0].split("For")[1])
        lst2.append(data["RatedFor"][i].split("Critically")[1].split("For")[1])
```

```python
data["Highly Rated For"] = lst1
```

```python
data["Critically Rated For"] = lst2
```

```python
data=data.drop(["RatedFor"],axis=1)
data
```

```python
for i in range(0,500):
    data["Parameters"][i]=data["Parameters"][i].replace(".","").replace("k","00")
data["Parameters"]
```

```python
lst3=[]
lst4=[]
lst5=[]
lst6=[]
for i in range(0,500):
    lst3.append(data["Parameters"][i].split("Reviews")[0].strip())
    lst4.append(data["Parameters"][i].split("Salaries")[0].split("Reviews")[1].strip())
    lst5.append(data["Parameters"][i].split("Interviews")[0].split("Salaries")[1].strip())
    lst6.append(data["Parameters"][i].split("Jobs")[0].split("Interviews")[1].strip())
```

```python
data["Reviews"],data["Salaries"],data["Interviews"],data["Jobs"]=lst3,lst4,lst5,lst6
```

```python
data=data.drop(["Parameters"],axis=1)
data
```

```python
data["Jobs"]=data["Jobs"].str.replace("--","0")
```

```python
data[["Reviews","Salaries","Interviews","Jobs"]] = data[["Reviews","Salaries","Interviews","Jobs"]].astype("int64")
```
